```python
import numpy as np
import matplotlib.pyplot as plt
from kafe2 import XYFit, Plot,  ContoursProfiler
#plt.style.use('dark_background')

abstand = np.array([0.6,0.61,0.62,0.63,0.635,0.64,0.645,0.65,0.66,0.67,0.68])
tfes = np.array([7.948,7.959,7.972,7.989,7.990,7.996,8.003,8.009,8.023,8.036,8.049])
tfes /= 5
tbew = np.array([8.211,8.134,8.071,8.018,7.995,7.975,7.958,7.940,7.912,7.892,7.877])
tbew /= 5

plt.plot(abstand, tfes, 'rp')
plt.plot(abstand, tbew, 'bp')
plt.ylabel('Periodendauer T [s]')
plt.xlabel('Abstand [m]')
plt.legend(['Periodendauer feste Schneide','Periodendauer bewegliche Schneide'])
plt.savefig('Pendel Versuch 1.2.pdf')
plt.show()

# Define the model function for the fit
def g(x, a, b):
    return a + b*x




yerr = 0.001
xerr = 0.001

fit = XYFit([abstand, tfes], model_function = g)
fit.add_error(axis = 'y', err_val = yerr)
fit.add_error(axis = 'x', err_val = xerr)
fit.do_fit()
plot = Plot(fit)
plot.plot()
plt.show()
fit.report()
plt.show()
```

```python
fit = XYFit([abstand[2:6], tbew[2:6]], model_function = g)
fit.add_error(axis = 'y', err_val = yerr)
fit.add_error(axis = 'x', err_val = xerr)
fit.do_fit()
plot = Plot(fit)
plot.plot()
fit.report()
plt.show()
```
