```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from kafe2 import XYFit, Plot,  ContoursProfiler

# Define the model function for the fit
def g(x, f, h):
    return -(f * (x + 1) + h)



# set data
y = 1000*np.ones(6) - np.array([865,860,850,840,830,800])
x = np.array([2.3,2.1,1.8,1.5,1.3,0.9])
yerr = 3
xerr = 0.03
# Define the fit
fit = XYFit(xy_data = [x,y], model_function = g)


# Add errors
fit.add_error(axis = 'y', err_val = yerr)
fit.add_error(axis = 'x', err_val = xerr)


# Do the fit
fit.do_fit()

# Finaly plot the fit
plot = Plot(fit)

plot.plot()

plt.savefig("Abbe10.pdf")
plt.show()
```
